```python
#MUESTRA LAS HORAS DE SOL ESTIMADAS PARA LA MISMA LOCALIZACION
#CALCULA LA MEDIA Y LA MÁXIMA
```

```python
days = response["features"][0]["properties"]["days"]
days
```

```python
for i in range(0, len(days),1):
    print(days[i]["variables"][0]["duration"])
```

```python
duracion_minutos = []
for i in range(0, len(days),1):
    hora = days[i]["variables"][0]["duration"].split("h ")
    minuto = hora[1].split("m")
    hora = int(hora[0])*60
    minuto = int(minuto[0])
    duracion_minutos.append(hora+minuto)

print(duracion_minutos)
```

```python
total = 0
for i in range(0, len(duracion_minutos),1):
    total += duracion_minutos[i]
print(total/len(duracion_minutos))
print(int((total/len(duracion_minutos))/60),"h ",int((total/len(duracion_minutos))%60),"m")
```

```python
max_duration = 0
for i in range(0, len(duracion_minutos),1):
   
    if(duracion_minutos[i]>max_duration):
        max_duration = duracion_minutos[i]

print(max_duration)
print(int(max_duration/60),"h ",int(max_duration%60),"m")
```

```python
#MUESTRA COMO UN DF LAS INFORMACION DE LAS MAREAS PARA HOY
```

```python
url = "https://servizos.meteogalicia.gal/apiv4/getTidesInfo"

response = requests.get(url,
                       params = {'API_KEY':key, "locationIds":"50500"})
response = response.json()
```

```python
mareas_hoy = response["features"][0]["properties"]["days"][0]["variables"][0]["summary"]

id_ = []
state = []
timeInstant = []

for i in range(0, len(mareas_hoy),1):
    
    id_.append(mareas_hoy[i]["id"])
    state.append(mareas_hoy[i]["state"])
    timeInstant.append(mareas_hoy[i]["timeInstant"])
    
df = pd.DataFrame(list(zip(id_, state, timeInstant)),columns =['id', 'Marea', 'Hora'])
df
```
